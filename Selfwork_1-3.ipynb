- считать movie_metadata.csv;
- перевести поле duration в часы;
- создать словарь {dtype: имя\_переменной};
- создать новый df, где индекс - это фамилия режиссера (предположим, что у режиссера фамилия всегда с позиции 1 в name.split(' ', 1));
- создать сводную таблицу: по каждому режиссеру вывести количество фильмов, средний и максимальный рейтинг на imdb, средний бюджет и количество фильмов снятых в 2012,2013,...,2015 годах

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movie_metadata.csv')
```

```python
df['duration'] = [i/60.0 for i in df['duration']]
```

```python
dict_types = {k:t for k,t in zip(df.keys(), df.dtypes)}
```

```python
df = df.dropna()
df['director_surname'] = [i.split(' ')[-1] if i.count(' ')>=1 else i for i in df['director_name']]
```

```python
df2 =df.groupby('director_surname', as_index=True)
```

```python
g = {}
for k,v in df2:
    g[k] = {
        'Num_of_films':len(v),
        'mean':v['imdb_score'].mean(),
        'max_IMDB':max(v['imdb_score']),
        'mean_budget':v['budget'].mean(),
        'films_2012-2015':len(list(filter(lambda x: x>=2012 and x<=2015, v['title_year'])))
        }
```

```python
pd.DataFrame(g).T
```
